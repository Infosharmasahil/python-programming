```python
#read a file that is tab separated 
products = pd.read_csv('products-2.csv', sep='\t')
```

```python
products
```

```python
#how many rows and columns
products.shape
```

```python
#inspect our data
products.head() #the first 5 rows
```

```python
#last five rows
products.tail()
```

```python
#summary stats
#basic statistical summary of all numeric values in the dataset
products.describe()
```

```python
#displaying subsets of our dataset
products['ListPrice'] #this returns a pandas series object
```

```python
products[['ProductID','ListPrice', 'Color', 'Weight']] #this returns a dataframe 
```

```python
#list the columns in the dataframe
products.columns
```

```python
#subets of our data based on certain conditions
#show all the products that are black (productID, Color, Weight)
products[products['Color'] == 'Black'][['ProductID','Color','Weight']]
```

```python
#list the productID, Name and ReorderPoint of all the products with a ListPrice greater than 500
products[products['ListPrice'] > 500][['ProductID','Name','ReorderPoint']]
```

```python
#check how many values there are of each type
products['Color'].value_counts()
```

```python
#how to create a new coloumn
products['DiscountPrice'] = products['ListPrice'] * 0.5
```

```python
products.head()
```

```python
#remove a column
products.drop(columns =['DiscountPrice'], inplace=True) 
```
