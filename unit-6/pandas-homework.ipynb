```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

```

```python
#read in the file 'Homicide.csv' and store it in a variable called homicide
homicide_df = pd.read_csv('Homicide.csv')
homicide_df
```

```python
#how many rows are in the dataset
homicide_df.shape[0]
```

```python
#list the names of all the columns in the dataset
homicide_df.columns
```

```python
#replace the default index with the values of the 'Index_' column
homicide_df.set_index(['Index_'])
```

```python
#list the types of homicides in the order of their frequency
homicide_df['Homicide_Type'].value_counts()
```

```python
#plot the homicide types and their frequencies on a bar chart


homicide_df['Homicide_Type'].value_counts().plot.bar()
```

```python
#find the year with the highest number of shootings

#homicide_df.groupby(['Homicide_Type']).max()

homicide_df[homicide_df['Homicide_Type'] == 'Shooting'].groupby('Occurrence_year')['Homicide_Type'].count().sort_values(ascending =False)
```

```python
#for 2018, list the months with the highest and lowest homicides
homicide_df['month'] = pd.DatetimeIndex(homicide_df['Occurrence_Date']).month
```

```python
def get_month_name(val):
    months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun','Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
    #error checking with try/except
    
    try:
        return months[val - 1]
    except:
        return val
```

```python
homicide_df['month'] = homicide_df['month'].apply(get_month_name)
```

```python
#list five safest neighborhoods
homicide_df.groupby('Neighbourhood')['Index_'].count().sort_values()[0:6]

```

```python
#which neighborhood had the highest number of stabbings
homicide_df[homicide_df['Homicide_Type'] == 'Stabbing'].groupby('Neighbourhood')['Homicide_Type'].count().sort_values(ascending=False)
```
